```python
import pandas as pd
```

```python
molm13 = pd.read_csv('GSE217195_counts_Molm13.txt', sep="\t") #read txt file
molm13
```

```python
#select columns
counts = molm13[["gene_id", "Sample_44527_C1_S00", "Sample_44528_C2_S00", "Sample_44529_91_S00", "Sample_44530_92_S00" ]]
counts = counts.set_index('gene_id')
counts
```

```python
#select columns if is not all zero
counts = counts[counts.sum(axis=1)>0]
counts
```

```python
counts = counts.T #transposed df 
counts
```

```python
#create your metadata
metadata = pd.DataFrame(zip(counts.index, ['WT','WT','KD','KD']), 
                        columns = ['Sample','Condition'])
metadata = metadata.set_index('Sample')

metadata
```

```python
from pydeseq2.dds import DeseqDataSet # %pip install pydeseq2
from pydeseq2.ds import DeseqStats

# Create a pyDESeq2 object
dds = DeseqDataSet(counts=counts,         #count data file .csv, txt, etc.
            metadata=metadata,            #samples what we used
            design_factors="Condition")   #sample condition labeled data

dds
```

```python
#deseq2(): This function typically performs the main analysis, calculating differential expression between conditions using the Negative Binomial GLM (Generalized Linear Model) approach implemented in DESeq2.
#The dds.deseq2() function would run the DESeq2 analysis on the dds object, generating results such as differential expression statistics, fold changes, p-values, and adjusted p-values.
dds.deseq2() 
```

```python
#get statistical result
stat_res = DeseqStats(dds, n_cpus=8 ,contrast = ('Condition', 'KD', 'WT'))
stat_res.summary()
```

```python
#result was turned to a df
result_df = stat_res.results_df
result_df
```

```python
import mygene #%pip install mygene

def add_gene_symbols_to_df(input_df):
    # Instantiate MyGeneInfo object
    mg = mygene.MyGeneInfo()
    
    # Retrieve unique gene IDs from the DataFrame index
    gene_ids = input_df.index.tolist()

    # Query for gene symbols using gene IDs
    results = mg.querymany(gene_ids, scopes='ensemblgene', species='human', fields='symbol', returnall=True)
    
    # Create a dictionary mapping gene IDs to gene symbols
    id_to_symbol = {entry['query']: entry.get('symbol') for entry in results['out']}

    # Map gene symbols to gene IDs in the DataFrame index
    input_df['Symbol'] = input_df.index.map(id_to_symbol)
    
    # Return the DataFrame with the added 'Symbol' column
    return input_df

# Example usage:
result_df_with_symbols = add_gene_symbols_to_df(result_df)

```

```python
#drop dublicate Symbol and keep first
result_df[result_df.duplicated(['Symbol'], keep='first')]

#drop 'None' value
result_df.dropna(subset=['Symbol'], how='all', inplace=True)

#select baseMean >=10 genes
result_df = result_df[result_df.baseMean >=10]
result_df
```

```python
import matplotlib.pylab as plt
import numpy as np

# Create a figure with a specific size
fig, ax = plt.subplots(figsize=(8, 6))

# Scatter plot
plt.scatter(x=result_df['log2FoldChange'], y=result_df['padj'].apply(lambda x: -np.log10(x)), s=1, label="Not significant", color="grey", marker=".")

# highlight down- or up- regulated genes
down = result_df[(result_df['log2FoldChange'] <= -1) & (result_df['padj'] <= 0.05)]
up = result_df[(result_df['log2FoldChange'] >= 1) & (result_df['padj'] <= 0.05)]

down_count = len(down)
up_count = len(up)

plt.scatter(x=down['log2FoldChange'], y=down['padj'].apply(lambda x: -np.log10(x)), s=20, label=f"Down-regulated ({down_count})", color="blue", marker="v")
plt.scatter(x=up['log2FoldChange'], y=up['padj'].apply(lambda x: -np.log10(x)), s=20, label=f"Up-regulated ({up_count})", color="red", marker="^")

# Axes labels and lines
plt.xlabel("$log_{2}$ fold change", size=10, weight='bold')
plt.ylabel("-$log_{10}$ FDR", size=10, weight='bold')

plt.axvline(-1, color="lightgrey", linestyle="--")
plt.axvline(1, color="lightgrey", linestyle="--")
plt.axhline(1, color="lightgrey", linestyle="--")

# Remove grid
plt.grid(False)

# Legend
legend_properties = {'weight': 'normal'}
plt.legend(fontsize=10, prop=legend_properties) 

# Title and display
plt.title("MOLM13 KD vs WT", weight='bold')

# Save plot with highest DPI
#plt.savefig('MOLM13_KDvsWT_plot.png', dpi=300)

plt.show()
```

```python
dds.layers['normed_counts'] #The 'normed_counts' layer likely contains these normalized expression counts for different genes across different samples.
dds.layers['log1p'] = np.log1p(dds.layers['normed_counts']) #This transformation is commonly applied to RNA-seq count data to stabilize variance and handle values close to zero effectively. The result is then assigned to a new layer named 'log1p' within the dds object.
dds
```

```python
dds_sigs = dds[:, significant.index] #This variable now contains the expression data (counts or other measurements) for only the significant genes identified in your analysis.
dds_sigs
```

```python
#In summary, the code snippet constructs a DataFrame named grapher from the expression data of significant genes stored in the dds_sigs object. 
grapher = pd.DataFrame(dds_sigs.layers["log1p"].T,
                      index=dds_sigs.var_names, columns=dds_sigs.obs_names)
grapher

```

```python
#ENSG to Symbol
grapher_with_symbols = add_gene_symbols_to_df(grapher)
# Set the 'Symbol' column as the index
grapher_with_symbols.set_index('Symbol', inplace=True)

grapher_with_symbols
    
```

When you set z_score=0, as you did in your sns.clustermap call, Seaborn standardizes the data along the rows, meaning it subtracts the mean and divides by the standard deviation for each row. This standardization ensures that each row (or column, depending on the z_score parameter) has a mean of 0 and a standard deviation of 1.

```python
import seaborn as sns

# Now, run clustermap with the updated DataFrame
#The resulting heatmap will show patterns of similarity and dissimilarity between rows and columns, with similar rows and columns grouped together based on hierarchical clustering.
ax = sns.clustermap(grapher_with_symbols, z_score=0, cmap='Spectral')

# Add a title to the color bar
cbar = ax.ax_heatmap.collections[0].colorbar
cbar.set_label('Z-score', rotation=-270, labelpad=20)

ax.fig.suptitle('MOLM13',  y=1.02, weight='bold')
```
